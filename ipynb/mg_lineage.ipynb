```python
umi_table = cas.pp.resolve_umi_sequence(
    umi_table,
    output_directory=output_directory,
    min_umi_per_cell=10,# 10
    min_avg_reads_per_umi=2.0,
    plot=False)
```

```python
umi_table
```

```python
umi_table3 = cas.pp.align_sequences(
    umi_table,
    ref_filepath='/media/chang/HDD-1/chang/weissman/PCT48-long.ref.fa',
    gap_open_penalty=20,
    gap_extend_penalty=1,
    n_threads=32,
)
```

```python
umi_table3 = cas.pp.call_alleles(
    umi_table3,
    ref_filepath='/media/chang/HDD-1/chang/weissman/PCT48-long.ref.fa',
    barcode_interval=(20, 34),
    cutsite_locations=[112, 166, 220],
    cutsite_width=12,
    context=True,
    context_size=5,
)
```

```python
umi_table3 = cas.pp.error_correct_umis(
    umi_table3,
    max_umi_distance=2,
    allow_allele_conflicts=False,
    n_threads=32,
)
```

```python
umi_table3.shape
```

```python
umi_table4 = cas.pp.filter_molecule_table(
    umi_table3,
    output_directory=output_directory,
    min_umi_per_cell=10, # 10
    min_avg_reads_per_umi=2.0,
    min_reads_per_umi=-1,
    intbc_prop_thresh=0.5,
    intbc_umi_thresh=10, # 10
    intbc_dist_thresh=1,
    doublet_threshold=None,
    allow_allele_conflicts=False,
    plot=False,
)
```

```python
umi_table4
```

```python
umi_table4.cellBC.unique().shape
```

```python
allele_table = cas.pp.call_lineage_groups(
    umi_table4,
    output_directory=output_directory,
    min_umi_per_cell=10, # 10
    min_avg_reads_per_umi=2.0,
    min_cluster_prop=0.005,
    min_intbc_thresh=0.05,
    inter_doublet_threshold=None,
    kinship_thresh=0.1,
    plot=True,
)
```

```python
allele_table
```

```python
df = pd.DataFrame(0, columns=pd.unique(umi_table4['cellBC']), index=pd.unique(umi_table4['allele']))
df
```

```python
umi = set()
for row in umi_table4.itertuples():
    if row.UMI not in umi:
        df.at[row.allele, row.cellBC] += 1
        umi.add(row.UMI)
df.to_csv("/media/chang/HDD-6/chang/mg/mg_lib_20220324/P4_1/allele_matrix.csv")
df
```

```python
df.sum(axis=0)
```

```python
rank_UMIs = pd.DataFrame(df.sum(axis=1), index = df.index, columns=df.sum(axis=0))
rank_UMIs['droplets'] = 'cell-free droplets'
rank_UMIs['droplets'] = rank_UMIs['droplets'].mask(rank_UMIs['total_counts']>50, 'droplet II')
rank_UMIs['droplets'] = rank_UMIs['droplets'].mask(rank_UMIs['total_counts']>100, 'droplet I')
rank_UMIs['droplets'] = rank_UMIs['droplets'].mask(rank_UMIs['total_counts']>500, 'cells')
rank_UMIs = rank_UMIs.sort_values(by='total_counts', ascending=False).reset_index().rename_axis("rank").reset_index()
rank_UMIs = rank_UMIs.loc[rank_UMIs['total_counts']>0]
rank_UMIs = rank_UMIs.set_index('index').rename_axis('cells')
```

```python
df.sum(axis=1)
```

```python
plt.figure(figsize=(4, 3), dpi=150)

ax = sns.lineplot(data = rank_UMIs,
                  x='rank',
                  y='total_counts',
                  hue='droplets',
                  hue_order=['cells', 'droplet I', 'droplet II', 'cell-free droplets'],
                  palette=sns.color_palette()[-4:],
                  markers=False,
                  lw=2)

ax.set_xscale('log')
ax.set_yscale('log')
ax.set_xlabel('sorted droplets');
ax.legend(loc='lower left', ncol=1, title=None, frameon=False)
ax.set_title(f'kneeplot: gw15_rep1');

sns.set_palette("muted")
sns.set_style("ticks")
sns.despine(offset=10, trim=False);
```

```python
bc = [line.rstrip() for line in open("/media/chang/HDD-1/chang/weissman/cell_bc.txt")]
```

```python
filtered_bc = df[set(bc).intersection(set(df.columns.to_list()))]
```

```python
filtered_bc
```

```python
diff = (set(bc) ^ set(df.columns.to_list())) & set(df.columns.to_list())
len(diff)
```

```python
empty_droplets = df[diff].T
empty_profile = empty_droplets.sum()/empty_droplets.sum().sum() 
```

```python
import scAR
```

```python
scarObj = scAR.model(raw_count=filtered_bc.T,
                     empty_profile=empty_profile.values,
                     scRNAseq_tech='CROPseq')

scarObj.train(epochs=400)
```

```python
scarObj.inference(feature_type='sgRNAs', cutoff=3)
```

```python
denoised_count = pd.DataFrame(scarObj.native_counts, index=filtered_bc.columns, columns=filtered_bc.index)
```
