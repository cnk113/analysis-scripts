```python
import numpy as np
import matplotlib.pyplot as plt
import vireoSNP
```

```python
res = vireoSNP.vcf.match_VCF_samples('/media/chang/HDD-3/chang/20220915_chimericorg_10x/cellranger/BMS753/vireo/GT_donors.vireo.vcf.gz', 
                                     '/media/chang/HDD-3/chang/20220915_chimericorg_10x/cellranger/CD2314/vireo/GT_donors.vireo.vcf.gz',
                                     GT_tag1 = 'PL', GT_tag2='PL')

fig = plt.figure()
vireoSNP.plot.heat_matrix(res['matched_GPb_diff'], 
                          res['matched_donors1'] , 
                          res['matched_donors2'] )
plt.title("Geno Prob Delta: %d SNPs" %(res['matched_n_var']))
plt.tight_layout()
plt.show()
```
