```python
import pandas as pd
from matplotlib import pyplot as plt
import datetime
```

```python
csv = pd.read_excel("~/coding_exercise_time_series_covid19_confirmed_global.xlsx")
csv.head()
```

```python
# Report the number of columns and rows in the dataset.
print("Rows: ", str(csv.shape[0]))
print("Cols: ", str(csv.shape[1]))
```

```python
# Report the number of unique Country/Regions represented in the dataset.
print("Unique country/regions: ", str(len(csv["Country/Region"].unique())))
```

```python
# Report the top 5 Country/Regions based on the number ovalue_countsince/States represented in the dataset. 
csv.groupby(["Province/State","Country/Region"]).size().reset_index()["Country/Region"].value_counts().head(5)
```

```python
# Report the top 5 and bottom 5 Country/Regions by the total number of confirmed cases on September 1, 2021.
csv.sort_values(by=datetime.datetime(2021,9,1))["Country/Region"].unique()[0:5]
```

```python
# Bottom 5
csv.sort_values(by=datetime.datetime(2021,9,1))["Country/Region"].tail(5)
```

```python
# Focus on Australia. Overlay scatter plots of the number of confirmed cases (y-axis) 
# for the first day of each month from February 2020 through September 2021 (x-axis), 
# with a different color for each Province/State. Use a log scale for the y-axis. Save the plot.
aus = csv[csv["Country/Region"] == "Australia"]
prov_cases = []
for i in range(2020,2022):
    if i == 2020:
        for j in range(2,13):
            prov_cases.append(aus[datetime.datetime(i,j,1)])
    if i == 2021:
        for j in range(1,10):
            prov_cases.append(aus[datetime.datetime(i,j,1)])
prov_cases = pd.DataFrame(prov_cases).T
prov_cases.index = aus["Province/State"]
prov_cases = prov_cases.T

plt.plot(prov_cases)
plt.xticks(rotation = 45)
plt.yscale('log')
plt.legend(prov_cases.T.index)
plt.title("Confirmed cases in Australia")
```

```python
# Plot a pie chart showing the distribution of confirmed cases for Australia on September 1, 2021, 
# where each wedge represents a Province/State. Donâ€™t worry about the pie chart looking pretty â€“ 
# several of the wedges are expected to be tiny. Save the plot.
plt.pie(prov_cases.T[datetime.datetime(2021,9,1)], labels=prov_cases.T.index)
```

```python
# Plot a bar chart showing the total number of cases by Country/Region for September 1, 2021 
# for the Country/Regions you listed in response to question 4. 
# Order the countries in the bar chart by their latitude,  
# such that the northernmost country is the first bar on the left. Save the plot. 
countries = csv.sort_values(by=datetime.datetime(2021,9,1))["Country/Region"].unique()[0:5]
top5 = csv.loc[csv["Country/Region"].isin(countries)]
top5 = top5.groupby(['Country/Region']).sum()
top5 = top5.sort_values(by="Lat",ascending=False)
plt.bar(top5.index, top5[datetime.datetime(2021,9,1)])
```
