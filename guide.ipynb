```python
from geomux import Geomux, read_table
import pandas as pd
import scanpy as sc
```

```python
opl33_lane1 = sc.read_10x_h5("/media/chang/HDD-1/chang/perturb/opl33_lane1/filtered_feature_bc_matrix.h5", gex_only=False)
opl33_lane2 = sc.read_10x_h5("/media/chang/HDD-1/chang/perturb/opl33_lane2/filtered_feature_bc_matrix.h5", gex_only=False)
opl34_lane1 = sc.read_10x_h5("/media/chang/HDD-1/chang/perturb/opl34_lane1/filtered_feature_bc_matrix.h5", gex_only=False)
opl34_lane2 = sc.read_10x_h5("/media/chang/HDD-1/chang/perturb/opl34_lane2/filtered_feature_bc_matrix.h5", gex_only=False)
```

```python
crispr_opl33_lane1 = opl33_lane1[:, opl33_lane1.var["feature_types"] == "CRISPR Guide Capture"].copy()
crispr_opl33_lane2 = opl33_lane2[:, opl33_lane2.var["feature_types"] == "CRISPR Guide Capture"].copy()
crispr_opl34_lane1 = opl34_lane1[:, opl34_lane1.var["feature_types"] == "CRISPR Guide Capture"].copy()
crispr_opl34_lane2 = opl34_lane2[:, opl34_lane2.var["feature_types"] == "CRISPR Guide Capture"].copy()
crispr = [crispr_opl33_lane1, crispr_opl33_lane2, crispr_opl34_lane1, crispr_opl34_lane2]
```

```python
geom = Geomux()
new = []
for i in crispr:
    i = i.to_df()
    i = pd.melt(i.reset_index(), id_vars='index')
    i = i.loc[i.value > 0,:]
    i.columns = ['barcode', 'guide', 'n_umi']
    i.n_umi = i.n_umi.astype(int)
    geom.fit(i)
    geom.predict()
    assignments = geom.assignments()
    new.append(assignments)
```

```python
new[0].to_csv('/media/chang/HDD-1/chang/perturb/opl33_lane1/geomux_guide_counts.tsv', sep='\t', index=False, header=True)
new[1].to_csv('/media/chang/HDD-1/chang/perturb/opl33_lane2/geomux_guide_counts.tsv', sep='\t', index=False, header=True)
new[2].to_csv('/media/chang/HDD-1/chang/perturb/opl34_lane1/geomux_guide_counts.tsv', sep='\t', index=False, header=True)
new[3].to_csv('/media/chang/HDD-1/chang/perturb/opl34_lane2/geomux_guide_counts.tsv', sep='\t', index=False, header=True)
```
